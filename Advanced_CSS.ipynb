**Problem Statement**

Explain the purpose of using the var() function in css. Also you have created two buttons with id named primaryBtn and secondaryBtn which should be given background colors using the var() function. The color code for primaryColor is #00b7ff and secondaryColor is #6c757d.

```python
index.html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>CSS var() Example</title>
  <link rel="stylesheet" href="styles.css">
</head>
<body>

  <button id="primaryBtn">Primary Button</button>
  <button id="secondaryBtn">Secondary Button</button>

</body>
</html>
```

```python
styles.css
:root {
  --primaryColor: #00b7ff;
  --secondaryColor: #6c757d;
}

#primaryBtn {
  background-color: var(--primaryColor);
  color: white;
  border: none;
  padding: 10px 16px;
  border-radius: 4px;
}

#secondaryBtn {
  background-color: var(--secondaryColor);
  color: white;
  border: none;
  padding: 10px 16px;
  border-radius: 4px;
}
```

Task 2:

**Problem Statement**

Create a 3D cube using the transform property of CSS.

```python
A.html
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <title>3D Cube</title>
  <link rel="stylesheet" href="A.css">
</head>
<body>

  <div class="scene">
    <div class="cube">
      <div class="face front">Front</div>
      <div class="face back">Back</div>
      <div class="face right">Right</div>
      <div class="face left">Left</div>
      <div class="face top">Top</div>
      <div class="face bottom">Bottom</div>
    </div>
  </div>

</body>
</html>
```

```python
A.css
body {
  display: flex;
  justify-content: center;
  align-items: center;
  height: 100vh;
  background: #f5f5f5;
}

/* Scene gives perspective */
.scene {
  width: 200px;
  height: 200px;
  perspective: 600px;
}

/* Cube */
.cube {
  width: 100%;
  height: 100%;
  position: relative;
  transform-style: preserve-3d;
  transform: rotateX(-20deg) rotateY(30deg);
}

/* Cube faces */
.face {
  position: absolute;
  width: 200px;
  height: 200px;
  background: rgba(0, 123, 255, 0.85);
  border: 2px solid #0056b3;
  color: #fff;
  font-size: 20px;
  font-weight: bold;
  display: flex;
  justify-content: center;
  align-items: center;
}

/* Individual faces */
.front  { transform: rotateY(0deg) translateZ(100px); }
.back   { transform: rotateY(180deg) translateZ(100px); }
.right  { transform: rotateY(90deg) translateZ(100px); }
.left   { transform: rotateY(-90deg) translateZ(100px); }
.top    { transform: rotateX(90deg) translateZ(100px); }
.bottom { transform: rotateX(-90deg) translateZ(100px); }
```

Task 3:

**Problem Statement**

Create a simple circular loader which will rotate continuously to look like a loading screen on a website

```python
P.html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <title>Loading Spinner</title>
    <link rel="stylesheet" href="P.css">
</head>
<body>

    <div class="loader-container">
        <div class="loader"></div>
    </div>

</body>
</html>
```

```python
P.css
/* Center the loader */
.loader-container {
    height: 100vh;
    display: flex;
    justify-content: center;
    align-items: center;
}

/* Circular loader */
.loader {
    width: 40px;
    height: 40px;
    border: 5px solid #e0e0e0;
    border-top: 5px solid #1e90ff;
    border-radius: 50%;
    animation: spin 1s linear infinite;
}

/* Animation */
@keyframes spin {
    0% {
        transform: rotate(0deg);
    }
    100% {
        transform: rotate(360deg);
    }
}
```
